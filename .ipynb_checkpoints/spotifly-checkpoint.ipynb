```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#import matplotlib.ticker as ticker
```

```python
%matplotlib inline
import os
```

```python
data = pd.read_csv("now_with_spotify.csv")
data.head(100)
```

```python
data = data.dropna()
data.head(100)
```

```python
danceable_data = data.sort_values(by='danceability', ascending=False)
danceable_data.head()
```

```python
sad_data = data.sort_values(by='acousticness', ascending=False)
sad_data.head(100)
```

```python
heat_data = data.drop(["spotify_id", "artist", "title", "key", "volume_number", "number"], axis=1)
```

```python
f,ax = plt.subplots(figsize=(15, 15))
sns.heatmap(heat_data.corr(), annot=True, linewidths=.7, fmt= '.1f',ax=ax)
```
